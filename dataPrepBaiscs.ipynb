```python
# Filtering & selecting data

#import numpy & pandas
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

```python
#Creating and labeling Indexes
series_obj = Series(np.arange(8), index=['row1', 'row2', 'row3', 'row4', 'row5', 'row6', 'row7', 'row8'])
series_obj
```

```python
#Integer Indexes
series_obj[-1]
series_obj[[0, 7]]
```

```python
# Creating a dataFrame Object
np.random.seed(25) # --> without this code. each time you run this code, you will get a different set of random values in the dataframe
DF_obj = DataFrame(np.random.rand(36).reshape((6,6)), index   =['row1', 'row2', 'row3', 'row4', 'row5', 'row6'],
                                                      columns =['column 1', 'column 2', 'column3', 'column 4', 'column 5', 'column 6'])
DF_obj
```

```python
# indexing by row & columns
DF_obj.loc[['row2', 'row5'], ['column 5', 'column 2']]
```

```python
# Data Slicing
series_obj[2:7]
```

```python
# Comparing with scalars value
# Scalar value is a single numerical value.
DF_obj > 0.2
```

```python
# Filtering with scalars
series_obj[series_obj > 2]
```
